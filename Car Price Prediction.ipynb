```python
import numpy as np
import pandas as pd
```

```python
# Loading the data

cars = pd.read_csv("CarPrice_Assignment.csv")

cars.head()
```

```python
# Removing the car model name from the column CarName

cars['CarName'] = [x.split(' ')[0] for x in cars['CarName']]

cars.head()
```

```python
# Looking for null values

cars.info()
```

```python
# Getting a summary of the data

cars.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Performing a correlation analysis on all the numerical variables

sns.pairplot(cars)
plt.show()
```

```python
# Correlation analysis on numerical variables

carsnums_1 = cars[["wheelbase","carlength","carwidth","carheight","curbweight", "enginesize","boreratio","stroke","compressionratio","horsepower","peakrpm","citympg","highwaympg"]]

sns.pairplot(carsnums_1)
plt.show()
```

```python
# Correlation analysis on closely associated variables. Based on the logic that all these variables are related to 
# physical dimensions of cars

carsnums_2 = cars[["wheelbase","carlength","carwidth","carheight","curbweight"]]

sns.pairplot(carsnums_2)
plt.show()
```

```python
# The previous correlation revealed that curbweight is the most useful variable and makes other physical dimension variables
# redundant. Correlation analysis on another set of closely associated variables. 

carsnums_3 = cars[["curbweight", "enginesize","boreratio","stroke","compressionratio","horsepower","peakrpm","citympg","highwaympg"]]

sns.pairplot(carsnums_3)
plt.show()
```

```python
# The previous correlation analysis revealed that both citympg and highwaympg are significantly positively correlated
# Hence it was logical to take mean of both and save it as a new coulmn "mpg".

cars["mpg"] = cars[["citympg", "highwaympg"]].mean(axis=1)

cars.head()
```

```python
# Analysing and manipulating numerical values
# Summary of previous correlation analyses ---
# 1. seems like curbweight is decently positively correlated with carwidth, carlength, wheelbase (carsnums_2)
# 2. enginesize and curbweight are also correlated (carsnums_3)
# 3. horsepower and enginesize are also correlated (carsnums_3)
# 4. citympg and highwaympg are perfectly correlated so it makes sense to average them and make a new mpg variable (carsnums_3)

# Dropping redundant variables - "carwidth","carlength","wheelbase","curbweight","horsepower","citympg","highwaympg"

cars["mpg"] = cars[["citympg", "highwaympg"]].mean(axis=1)

cars = cars.drop(["carwidth","carlength","wheelbase","curbweight","horsepower","citympg","highwaympg"], axis = 1)

cars.head()
```

```python
# Checking out the data visually

plt.figure(figsize=(20, 12))
plt.subplot(2,3,1)
sns.boxplot(x = 'fueltype', y = 'price', data = cars)
plt.subplot(2,3,2)
sns.boxplot(x = 'aspiration', y = 'price', data = cars)
plt.subplot(2,3,3)
sns.boxplot(x = 'doornumber', y = 'price', data = cars)
plt.subplot(2,3,4)
sns.boxplot(x = 'carbody', y = 'price', data = cars)
plt.subplot(2,3,5)
sns.boxplot(x = 'drivewheel', y = 'price', data = cars)
plt.subplot(2,3,6)
sns.boxplot(x = 'enginelocation', y = 'price', data = cars)
plt.show()
```

```python
# Checking out the data visually

plt.figure(figsize=(20, 12))
plt.subplot(2,3,1)
sns.boxplot(x = 'enginetype', y = 'price', data = cars)
plt.subplot(2,3,2)
sns.boxplot(x = 'cylindernumber', y = 'price', data = cars)
plt.subplot(2,3,3)
sns.boxplot(x = 'fuelsystem', y = 'price', data = cars)
plt.show()
```

```python
# Converting categorical variable "cylindernumber" to numerical. 

cars["cylindernumber"] = cars["cylindernumber"].replace({'two': 2, "three": 3, "four": 4, "five": 5, "six": 6, "eight": 8, "twelve": 12})

cars.head()
```

```python
# Converting the superficial numerical variable's datatype to actually be numerical in nature and performing a correlation 
# between cylindernumber and engine size.

cars['cylindernumber'] = cars.cylindernumber.astype(float)
carsvar = cars[['cylindernumber','enginesize']]
sns.pairplot(carsvar)
plt.show()
```

```python
# Cylindernumber is correlated with engine size and so cylindernumber is redundant.

cars = cars.drop(["cylindernumber"], axis=1)
```

```python
# Converting catergorical enginetype to numerical.

cars["enginetype"] = cars["enginetype"].replace({"dohc": 0, "ohcv": 1, "ohc": 2, "l": 3, "rotor": 4, "ohcf": 5, "dohcv": 6})
```

```python
# Checking to see potential redundancies. Since there is not correlation, keeping enginetype.

carsvar_2 = cars[['enginetype','enginesize']]
sns.pairplot(carsvar_2)
plt.show()
```

```python
# Converting drivewheel to numerical.

cars["drivewheel"] = cars["drivewheel"].replace({"rwd": 0, "fwd": 1, "4wd": 2})
```

```python
# Converting fuelsystem to numerical

cars["fuelsystem"] = cars["fuelsystem"].replace({"mpfi": 0, "2bbl": 1, "mfi": 2,"1bbl": 3, "spfi": 4, "4bbl": 5,"idi": 6, "spdi": 7})
```

```python
# Converting carbody to numerical.

cars["carbody"] = cars["carbody"].replace({"convertible": 0, "hatchback": 1, "sedan": 2,"wagon": 3, "hardtop": 4})
```

```python
# Dropping variables that didn't seem a determinant of car price.

cars = cars.drop(["fueltype", "aspiration", "doornumber", "enginelocation","carheight", "car_ID","CarName"], axis=1)
```

```python
cars.head()
```

```python
# Pefroming a train-test split

from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train, df_test = train_test_split(cars, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
num_vars = ["symboling","enginesize","boreratio","stroke","compressionratio","peakrpm","mpg","price"]

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
# Mesauing variable correlations 

plt.figure(figsize = (16, 10))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
# Removing price.

y_train = df_train.pop('price')
X_train = df_train
```

```python
import statsmodels.api as sm

X_train_lm = sm.add_constant(X_train[['enginesize']])

lr = sm.OLS(y_train, X_train_lm).fit()

lr.params
```

```python
# Visualising the data with a scatter plot and fitting a regression line.

plt.scatter(X_train_lm.iloc[:, 1], y_train)
plt.plot(X_train_lm.iloc[:, 1], -0.072888 + 1.210676*X_train_lm.iloc[:, 1], 'r')
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# I dropped variables one by one and after assessing only removed these at once.
# This stage was reached only after methodically removing one variable at a time.

X = X_train.drop(["stroke", "carbody", "symboling","boreratio", "fuelsystem", "drivewheel"], 1)
```

```python
# The VIFs and p-values are within appropriate ranges.

vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X_train_lm = sm.add_constant(X)

lr_2 = sm.OLS(y_train, X_train_lm).fit()
```

```python
print(lr_2.summary())
```

```python
y_train_price = lr_2.predict(X_train_lm)
```

```python
# Plotting the histogram of the error terms.

fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  
plt.xlabel('Errors', fontsize = 18)  
```
